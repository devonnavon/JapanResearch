```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 

sns.set()
```

<h3>Import Data</h3>

```python
population = pd.read_csv('world-bank-data/Population/data.csv', header=2)
migrant_pop = pd.read_csv('world-bank-data/MigrantPop/data.csv', header=2)
gdp = pd.read_csv('world-bank-data/GDP/data.csv', header = 2)
gni = pd.read_csv('world-bank-data/GNI/data.csv', header = 2)
```

```python
focus = ['Japan', 'United States', 'Australia', 'United Kingdom', 'Korea, Rep.',] #'China', 'Singapore']
drop = ['Country Code', 'Indicator Name', 'Indicator Code', 'Unnamed: 64']

#function to index and drop
fc = lambda df: df[df['Country Name'].isin(focus)].drop(drop, axis=1).set_index('Country Name', drop=True).T
idx = lambda df: df.set_index(df.index.astype(int).rename('Year'))
form = lambda df: idx(fc(df))
```

```python
full = form(population)
lst_20 = full[full.index>2000]
lst_5 =full[full.index>2014]
```

```python
plt.figure(figsize=(21, 7))


plt.subplot(131)
plt.plot(full)
plt.ylabel('Population') 
plt.title('full history')

plt.subplot(132)
plt.plot(lst_20)
plt.title('20 year')

plt.subplot(133)
plt.plot(lst_5)
plt.title('5 year')


plt.suptitle('Population Growth')
plt.show()
```
