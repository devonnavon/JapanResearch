<h3>Import Data</h3>

```python
population = wb.transform(pd.read_csv('world-bank-data/Population/data.csv', header=2))
migrant_pop = wb.transform(pd.read_csv('world-bank-data/MigrantPop/data.csv', header=2))
gdp = wb.transform(pd.read_csv('world-bank-data/GDP/data.csv', header = 2))
gni = pd.read_csv('world-bank-data/GNI/data.csv', header = 2)#lots of info is it worth it?
```

```python
focus = ['Japan','Singapore', 'United Kingdom', 'China', 'Korea, Rep.','United States']

stats = [population, migrant_pop, gdp]
all_vals = pd.concat(stats, sort=True).reset_index()
all_vals = all_vals[all_vals['Country Name'].isin(focus)]
all_vals.columns
```

```python
fig = px.line(all_vals, x = 'Year',  y = 'GDP (current US$)', color='Country Name')
```

```python
fig.show()
```

```python
mp = migrant_pop.reset_index()
mp = mp[mp['Country Name'].isin(focus)]
```

```python
fig = px.line(mp, x = 'Year',  y = 'International migrant stock, total', color='Country Name')
fig.show()
```
